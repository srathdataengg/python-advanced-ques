# Question 3: Maximum Number of Vowels in a Substring (LC #1456)

The Scenario: You are looking at a string of text.
You want to find a section of length $k$ that contains the most vowels ($a, e, i, o, u$).

Input: s = "abciiidef", k = 3

Window 1: "abc" -> 1 vowel ('a')
Window 2: "bci" -> 1 vowel ('i')
Window 3: "cii" -> 2 vowels ('i', 'i')Window 4: "iii" -> 3 vowels (Max!)Output: 3

```python
def max_vowels(s:str,k:int):
    """ :return - maximum numbers of vowels a substring holds"""
    vowels = {"a","e","i","o","u"}
    current_count = 0
    max_count = current_count

    #1. Build the first window
    for i in range(k):
        if s[i] in vowels:
            current_count +=1
    # 2. Slide the Window from k to the end
    for i in range(k,len(s)):
        # Entering : The new character at index i
        if s[i] in vowels:
            current_count +=1
        # Leaving: The old character at index at (i-k) why i-k because our new position is +k i.e i so i-k will remove -3(here k=3) position
        char_leaving = s[i-k]
        if char_leaving in vowels:
            current_count = current_count -1

        # Update our record
        if current_count > max_count:
            max_count = current_count

    return max_count

print(max_vowels(s="abcdeiiiiefg",k=3))



```

# Question 5. Maximum Average Subarray I (LC #643)

• Goal: Given an array `nums` and integer `k`, find the contiguous subarray of length `k` that has the maximum average value.

• Key Logic: Use a fixed window of size `k`. Calculate the initial sum, then slide by adding the new element and subtracting the one leaving.

• Input: `nums = [1, 12, -5, -6, 50, 3], k = 4`

• Output: `12.75`

```python
def find_max_average(nums,k):
    """ let's find the avergae of first k elements then slide forward check which is max-average k elements"""
    current_avg_sum = 0
    max_avg_sum = 0

    # Step-1 Initial window of k elements
    for i in range(k):
        current_avg_sum = (current_avg_sum + nums[i])/k

    max_avg_sum = current_avg_sum

    # Step-2 Sliding the window
    # We start with kth point and move forwards and then deduct -k position
    for i in range(k,len(nums)):
        current_avg_sum = (current_avg_sum + nums[i] + nums[i-k])/k

        if current_avg_sum > max_avg_sum:
            max_avg_sum = current_avg_sum

    return max_avg_sum

# Test Case
nums = [1, 12, -5, -6, 50, 3]
k = 4
print(find_max_average(nums, k)) # Should be 12.75

```
