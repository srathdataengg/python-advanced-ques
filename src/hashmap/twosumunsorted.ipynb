1. Two Sum (LC #1)
Problem Statement: Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target. You may assume that each input would have exactly one solution, and you may not use the same element twice.

Example:

Input: nums = [2, 7, 11, 15], target = 9

Output: [0, 1] (Because nums[0] + nums[1] == 9)

```python
from typing import List


def two_sum(nums:List[int],target:int):
    seen = {}
    for i, num in enumerate(nums):
        needed = target - num
        if needed in seen:
            return [seen[needed],i]
        seen[num] = i

    return []

print(two_sum([2,7,11,15],target=9))
```


### _**Question 4: **Intersection of Two Arrays II (LC #350)****_

The Scenario (CrowdStrike Context): Imagine you have two lists of "Threat IDs."

List 1: Threats detected by the Firewall.

List 2: Threats detected by the Antivirus. You want to find out which threats appear in both lists. If a threat appears twice in both, you should report it twice.

The Problem: Given two integer arrays nums1 and nums2, return an array of their intersection.

Sample Input 1: nums1 = [1, 2, 2, 1], nums2 = [2, 2]

Sample Output 1: [2, 2]

Sample Input 2: nums1 = [4, 9, 5], nums2 = [9, 4, 9, 8, 4]

Sample Output 2: [4, 9] (or [9, 4])

The Strategy: "The Grocery List"
Step 1: Loop through nums1 and count how many times each number appears. Store this in a dictionary (our "Inventory").

Step 2: Loop through nums2. For every number you see, check your "Inventory."

Step 3: If the number is in the inventory and the count is greater than 0:

Add it to your result list.

Crucial Step: Subtract 1 from the inventory count (because you've "used up" one match).

```python
def intersect(nums1,nums2):
    counts = {}
    for i in nums1:
        if i in counts:
            counts[i] +=1
        else:
            counts[i] = 1
    print(counts)
    result = []
    for num in nums2:
        if num in counts and counts[num] > 0:
            result.append(num)
            counts[num] -=1
            print(counts)
    return result

print(intersect(nums1=[2,2],nums2=[2,2,2,2,2,2]))
```
